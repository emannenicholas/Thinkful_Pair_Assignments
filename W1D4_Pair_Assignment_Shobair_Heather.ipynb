```python
import numpy as np
import matplotlib.pyplot as plt

dataset = np.random.poisson(5, 1000000)
master_mean = dataset.mean()
dataplot = plt.hist(dataset, bins=10)
master_mean
```

```python
#small subset
setsize = 100
subset = np.random.choice(dataset, setsize)
submean = subset.mean()
subplot = plt.hist(subset)
submean
```

```python
#medium subset
setsize = 5000
subset = np.random.choice(dataset, setsize)
submean = subset.mean()
subplot = plt.hist(subset)
submean
```

```python
#large subset
setsize = 10000
subset = np.random.choice(dataset, setsize)
submean = subset.mean()
subplot = plt.hist(subset)
submean
```

```python
#X-large subset
setsize = 100000
subset = np.random.choice(dataset, setsize)
submean = subset.mean()
subplot = plt.hist(subset)
submean
```

```python
#tertiary subset
tertsetsize = 1000
tertset = np.random.choice(subset, tertsetsize)
tertmean = tertset.mean()
tertplot = plt.hist(tertset)
tertmean
```

```python
tertmeandist = []
tertsetsize = 1000
for i in range(10000):
    tertset = np.random.choice(subset, tertsetsize)
    tertmean = tertset.mean()
    tertmeandist.append(tertmean)

tertmeanplot = plt.hist(tertmeandist)
tertmeanmean = np.mean(tertmeandist)
print(tertmeanmean)
print(master_mean)
```
